```python
# import pandas library to read csv into DataFrame
import pandas as pd
# import py2neo library to connect and load data to Neo4j Graph
from py2neo import Graph, Node, Relationship
# import networkx, matplotlib for visualization
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#declare a graph variable which connects to Neo4j Graph
graph = Graph("bolt://localhost:7687", auth=("neo4j", "password"))
#load nodes and relationships to DataFrame
df_rel = pd.DataFrame(graph.run("MATCH (a)-[]->(b) RETURN a, b").data())
df_node = pd.DataFrame(graph.run("MATCH (n:airport) RETURN n").data())
```

```python
#networkx directed graph
G = nx.DiGraph()
#color and size of node depends on its rank
colors = []
sizes=[]
#add nodes to graph G and values of colors and sizes.
for index, row in df_node.iterrows():
    rank = row['n']['rank']
    G.add_node(row['n']['name'], rank=rank)
    colors.append(rank)
    sizes.append(rank*1000)
#add edges to graph G
for index, row in df_rel.iterrows():
    G.add_edge(row['a']['name'], row['b']['name'])
# visualization
plt.figure(1,figsize=(12,12))
nx.draw(G, node_color= colors ,with_labels=True, node_size=sizes,font_size=10, font_color='red')
plt.show()
```
